```python
# Vemos la tabla y nos fijamos dónde están los datos
df = pd.read_html('https://www.promiedos.com.ar/', match='COPA DE LA LIGA')
df = df[0]
df
```

```python
# Creamos un Dataframe vacío para ir poblándolo con la tabla scrapeada, pero antes parseada
df_partidos = pd.DataFrame(columns=['equipo_local',
                                    'equipo_visitante',
                                    'goles_local',
                                    'goles_visitante',
                                    'torneo',
                                    'goleadores_local',
                                    'goleadores_visitante',
                                    'estado_partido',
                                    'detalle'])

df_partidos
```

```python
# Vemos otra liga diferente
df = pd.read_html('https://www.promiedos.com.ar/', match='LA LIGA')
df = df[-1]
df
```

```python
# Referenciamos los datos a variables para ir ordenándolos y plasmarlos en filas
# Algunos tienen el detalle debajo de la fila que describe la liga, por ej: copa de la liga -> cuartos de final visto más arriba

equipo_local = df.iloc[2, 1]
goles_local = df.iloc[2, 2]
goles_visitante = df.iloc[2, 3]
equipo_visitante = df.iloc[2, 4]
torneo = df.iloc[1, 0]
goleadores_local = df.iloc[3, 1]
goleadores_visitante = df.iloc[3, 4]

print('Equipo local: ' + str(equipo_local))
print('Goles local: ' + str(goles_local))
print('Goles visitante: ' + str(goles_visitante))
print('Equipo visitante: ' + str(equipo_visitante))
print('Torneo: ' + str(torneo))
print('Goleadores local: ' + str(goleadores_local))
print('Goleadores visitante: ' + str(goleadores_visitante))
```
